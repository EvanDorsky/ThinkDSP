```python
from __future__ import print_function, division

import numpy as np
import thinkdsp
import thinkplot
import copy
import ps01
```

```python
hairdryer = thinkdsp.read_wave('ps01/sample.wav')
```

```python
start = 4
duration = .6
segment = hairdryer.segment(start, duration)
segment.make_audio()
```

```python
spectrum = segment.make_spectrum()
lowspec = copy.deepcopy(spectrum)
lowspec.plot()
```

```python
lowspec.low_pass(1000)
lowspec.plot()
```

```python
lowwave = lowspec.make_wave()
lowwave.make_audio()
```

```python
lowwave.plot()
highspec = copy.deepcopy(spectrum)
highspec.high_pass(5000)
```

```python
highspec.plot()
```

```python
highwave = highspec.make_wave()
highwave.make_audio()
```

```python
highwave.plot()
```

```python
bandspec = copy.deepcopy(spectrum)
bandspec.band_stop(0, 600)
bandspec.low_pass(1000)
bandspec.plot()
```

```python
bandwave = bandspec.make_wave()
bandwave.make_audio()
```

```python
bandwave.plot()
thinkplot.pyplot.xlim([.1, .14])
```

Function with even+odd harmonics and $\dfrac{1}{f^2}$ falloff?

```python
coS = thinkdsp.CosSignal

fundamental = 440

freq = fundamental
mystery = coS(freq=freq, amp=1/(freq**2))
for i in range(30):
    freq += fundamental
    mystery += coS(freq=freq, amp=1/(freq**2))
    
mystery.make_wave().make_spectrum().plot()
```

```python
mystery.make_wave().make_audio()
```

```python
mystery.make_wave().plot()
thinkplot.pyplot.xlim([0, .02])
```
