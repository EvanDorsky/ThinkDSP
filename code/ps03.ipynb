```python
import thinkdsp, pandas, thinkplot, numpy

nrows = 1625
df = pandas.read_csv('ps03/coindesk-bpi-USD-close.csv', nrows=nrows, parse_dates=[0])
ys = df.Close.values
wave = thinkdsp.Wave(ys, framerate=1)
wave.plot()
```

## 1) Compute the power spectrum of BitCoin prices. Does it resemble white, pink, or Brownian noise?

```python
spec = wave.make_spectrum().plot_power()
thinkdsp.pyplot.xlim([0, 0.01])
```

It looks like brownian noise because most of the power is concentrated in the lowest frequencies and there's a pretty steep falloff.

## 2) Compute the autocorrelation of BitCoin prices. Does the autocorrelation function decay exponentially? Is there evidence of periodic behavior?

```python
def serial_corr(wave, lag=1):
    n = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:n-lag]
    corr = numpy.corrcoef(y1, y2)[0, 1]
    return corr
    
def autocorr(wave):
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs

lags, corrs = autocorr(wave)
thinkplot.plot(lags, corrs)
```

There appears to be some periodic behavior because the autocorrelation has periodic bumps. That's pretty cool. Let's plot a semilog x plot to see if it falls off exponentially.

```python
thinkdsp.pyplot.semilogx(lags, corrs)
```

The falloff isn't really a straight line on the semilogx plot above, which confirms my thought that it's not pink noise.

## 5) ... Implement a version of this algorithm and apply it to a recording of music or speech.

```python
def compress(wave, thresh):
    fr = wave.framerate # good to know
    seg_len = 10000.0 # each segment has this many samples
    ## 1. Break a long signal into segments
    segs = []
    for i in range(int(len(wave)/seg_len) -1): # cheating with -1
        portion = seg_len/fr # useful ratio
        pos = i*portion # position in the wave (seconds)
        segs.append(wave.segment(start=pos, duration=portion))
        
    ## 2. Compute the DCT of each segment
    dcts = [seg.make_dct() for seg in segs]
    
    ## 3. Identify frequency components with amplitudes
    ## so low they are inaudible, and remove them.
    ## Store only the frequencies and amplitudes that remain.
    return [thinkdsp.Dct(map(lambda a: a if abs(a) > thresh else 0,
                         dct.amps), fr) for dct in dcts]
    
    ## 4. To play back the signal, load the frequencies and amplitudes
    ## for each segment and apply the inverse DCT.
    
def uncompress(compressed):
    fr = compressed[0].framerate
    return thinkdsp.Wave(
        numpy.array(map(lambda dct: dct.make_wave().ys,
                        compressed))
        .flatten(), fr)

sax.apodize()
uncompress(compress(sax, 6000)).make_audio()
```
